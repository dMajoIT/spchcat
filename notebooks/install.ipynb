## Fetch the Package

We'll install the `spchcat` speech recognition tool from a `.deb` package, downloaded from [github.com/petewarden/spchcat/releases](https://github.com/petewarden/spchcat/releases), so use `wget` to fetch the latest version.

```python
!wget -q https://github.com/petewarden/spchcat/releases/download/v0.0.2-alpha/spchcat_0.0-2_amd64.deb
```

## Install Dependencies

We do need the `sox` library, so make sure it's installed.

```python
!apt-get -q install -y sox
```

## Install the Package

On a desktop machine we could just double-click on the package to open and install it, but because we're in a notebook we'll use the `dpkg` tool instead.

```python
!dpkg -i spchcat_0.0-1_amd64.deb
```

## Test `spchcat`

The tool needs an audio input to run on, and since this may be running in Colab or another environment without a microphone attached, we'll download some test audio and run it on that.

```python
!wget --quiet https://github.com/coqui-ai/STT/releases/download/v1.1.0/audio-1.1.0.tar.gz
!tar -xzf audio-1.1.0.tar.gz
```

```python
!spchcat audio/4507-16021-0012.wav
```
